```python

import pandas as pd
import os
from datetime import datetime


path = 'arquivo.xlsx'
data = pd.read_excel(path, header=None)
data['Tarefao'] = '0'
data['Tarefinha'] = '0'
projectName = data.columns[1]

metaData = data.iloc[0:7, 0:2].copy()
metaData = metaData.T
metaData.columns = metaData.iloc[0, :]
metaData.index = [0, projectName]
metaData.drop([0], inplace=True)

colunas = list(data.iloc[8, 0:-2].values)
colunas.append('Tarefao')
colunas.append('Tarefinha')
data.drop(range(0, 9), inplace=True)
data.columns = colunas


col_ref = 1
ultimaTarefona = data.iloc[0, col_ref]
for i in range(len(data)-1):
    if len(data.iloc[i, col_ref].split('.')) < len(data.iloc[i+1, col_ref].split('.')):
        ultimaTarefona = data.iloc[i, col_ref]
        data.iloc[i, 16] = ultimaTarefona
    else:
        data.iloc[i, 17] = ultimaTarefona

countLines = len(data)-1
if len(data.iloc[countLines, col_ref].split('.')) < len(data.iloc[countLines-1, col_ref].split('.')):
    data.iloc[countLines,16] = data.iloc[countLines, col_ref]
else:
    data.iloc[countLines, 17] = ultimaTarefona
```
